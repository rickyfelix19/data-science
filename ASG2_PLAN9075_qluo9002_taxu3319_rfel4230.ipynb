### Background

There are a lot of reason why people travel to Australia, it might be for studying reasons, business, travelling, or as a mean to settle down to the "the Great Southern Land". For many international, it is well known to be one of the most favorable places for skilled migrants to come and to live a better life. Australia is also considered to be the world's major 'immigration nations'. And since 1945 (up to 2017), there are over 7.5 Million people have settled here and Australia's overseas-born resident population—estimated to be 28.2 per cent of the population in June 2015. (Phillips & Simon-Davies, 2017) The Skilled Visa to Australia is for skillful individuals or families to move permanently to Australia, filling Australia’s skill shortages. There are a couple of ways that a person or a family could receive said visa, it could be sponsored by themselves, state, family, and employer. (Australia Skilled Migration, 2021). This leads to Population Growth and according to data pattern, the net overseas migration population in Australia has always increased each year unless it is due to unprecedented events, such as: World War, The Great Depression, Covid-19 Pandemic (Australia Bureau Statistics, 2021)

### Problems:

#### 1. Congestion & Overpopulation
> One of the current problems within the Australia could be the overpopulation of the mega cities and there’s a cleavage between the middle and small cities. Middle cities are absent but dwell in either mega cities or small cities and this reflects the problem of polarization of population distribution. Australian government made policies such as visa 491 and visa 190 to encourage the immigrants to live within the regional area. The principle “Nomination or Sponsorship” is to support the decentralization of population and urging to develop the regional areas. 

#### 2. Migrants Policy
> According to the policy change of increased skilled worker wage threshold and the increase in the quota of Permanent Residency, it could be distinguished that the government of Australia is intending to distribute and encourage more immigrants to stay and develop the regional area. However, according to the statistics on the residency of new immigrants, abundant of immigrants especially skilled workers moved to mega cities after they acquired permanent residency or immigrated. Therefore, we were intrigued by the question of why regional area are not attractive to not only the immigrants to stay, but also native residents, and what factors might affect this phenomenon? 

The objective is to increase better life in the regional areas by using indicators based on Placemaking, Megacities, and the influence of someone moving away

## Policy Reflection

The current visa policy for overseas students has been extended from 3 years to 5 years (for master’s degree) which matched the process of getting VETASSESS certification, acquiring working experience and the waiting time for being invited. This new action is already enhanced the chance that retaining the educated overseas students within Australia. This report assesses the rent and income as the main indicators and marks its influence by analyzing the existing phenomenon. According to the analysis, one of the barriers to retaining them within the regional area might be the shortage of job opportunities. As the diagram OF "ICT Professions" shows, most of the regional area are rarely have the related skill workers and at the same time they are clustering in the regional area that surrounding to the mega cities for example Yass which is adjacent to Canberra. If the immigrant need further career progression, they might not have choice and moving to the mega cities might be the only solution, as the regional area do not have related large enterprises. However, the analysis also shows that the professionals are starting to retain in the surrounding of the mega cities, it could be argued that the decentralization of population moved a concrete step. It could be also perceived that NSW is gradually suburbanizing and on the stage that initial suburbanization stage (Liu, Sun & Peng 2022). Due to the limitation of time, this report does not have the opportunity to assess the migrant movement across the time, which could take into the consideration in the future analysis in order to assess the influence international immigration and domestic migration policies. 

Therefore, policies that encourage entrepreneurship in the regional area such incentive to businesses and Incubator Support Initiatives might be the key to provide employment opportunities and hence foster the development in the regional area. Special supporting to the shortage services is suggested by this report in terms of funding and expert-in-residency (EIR), by incubating various type of businesses in the regional area, it is more likely to retain the migrants to come and stay within the regional area.

Another perspective is that: as being observed through the data, some interesting phenomenon is found, for example Yass and Goulburn again, is predominantly filled with ICT professionals while the surroundings might rarely have this genre of skilled workers. A healthy structure of a town or regional area should be in a multi-diverse model rather than being monopoly by one single job. It is important to sparse the employment opportunities across different regional areas in order to distribute various types of skilled workers to the towns in order to provide a comprehensive service to the area (Cohen et al. 2015). By providing a more comprehensive service, the amenities within the region would be improved therefore improving the overall image to the migrants. Due to the time limit, this report is not able to cover the analysis on influence of the city image and infrastructures on the migrations. In the future research, by incorporating the indicators analyzed by this report with the analysis of living quality such as green canopy, quantity of infrastructure, it could make a further comprehensive analysis on the immigrate behaviors.

The diagram titled "Rent Affordability" indicates that rent stress is not a significant concern. However, the affordability of rent could raise questions about the quality of the housing provided. This report acknowledges a limitation in its scope, as it does not encompass a comprehensive investigation and analysis of housing and environmental quality. It would be prudent for future analyses to incorporate assessments of these factors. Enhanced housing and living environment standards can notably improve the quality of life. As the diagram indicates, the suburb with higher median income is more likely to have a higher rent cost. This might be due to the demand of high-income populations that stimulate the improvement of the housing quality. This, in turn, is a potent strategy for attracting and retaining immigrants, contributing to the area's sustained growth and diversity.

```python
# Merge SA2 with INCP

df_sa2 = pd.read_csv('incp.csv')
df_sa2.rename(columns = {'INCP Total Personal Income (weekly) / SA2': 'SA2'},
              inplace = True)
print(df_sa2.shape)
df_sa2
```
